### Import Dependencies

```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
#Read in the csv file
all_crash_file = "../Resources/Crash_Information_(Last_5_Years).csv"
all_crash_df = pd.read_csv(all_crash_file)
all_crash_df.head()
```

### Transform all_crash DataFrame

```python
# Create a filtered dataframe from specific columns
allcrash_df = ["OBJECTID", "ACC_ID", "LONGITUDE", "LATITUDE", "CRASH_DATE", "CRASH_TIME", "SEVERITY"]
allcrash_transformed= all_crash_df[allcrash_df].copy()
allcrash_transformed.head()
```

```python
#Filter to have only FATAL data
allcrash_fatal = allcrash_transformed.loc[allcrash_transformed['SEVERITY'] == "Fatal"]
allcrash_fatal
```

```python
#Use list comprehension to change uppercase letters to lowercase
[x.lower() for x in allcrash_fatal.columns]
```

```python
#Put new values into dataframe
allcrash_fatal.columns = [x.lower() for x in allcrash_fatal.columns]
allcrash_fatal
```

### Save cleaned dataframe as csv file

```python
allcrash_fatal.to_csv('Resources/Crash_Information_cleaned.csv')
```
